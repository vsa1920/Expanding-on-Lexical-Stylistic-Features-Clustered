# Extension - clustered anisotropy removal
Instead of globally removing the anisotropic dimensions, we now cluster them and remove anistropy from each cluster independently.

## Clustered Anistropy removal with larger models

Larger models seem to be more anisotropic, clustering them results in outlier clusters. Let us think of a way to deal with them.

One way is to totally disregard, all the outliers. We remove the outliers by removing clusters with less than 10 points when calculating the PCs, mean and standard deviation.

```python
!python predict.py --feature "complexity" --dataset_name "SimplePPDB" --split "val" --model_name "xlm-roberta-large_4_layeragg_seeds7_max_normalized"
```
